```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('recruitment_data.csv')

# Calculate hiring rate
hiring_rate = data['HiringDecision'].mean() * 100

# Plotting the hiring rate
plt.figure(figsize=(8, 5))
plt.bar(['Hiring Rate'], [hiring_rate], color='skyblue')
plt.ylim(0, 100)
plt.title('Hiring Rate')
plt.ylabel('Percentage (%)')
plt.axhline(y=hiring_rate, color='r', linestyle='--', label=f'Average Hiring Rate: {hiring_rate:.2f}%')
plt.legend()
plt.show()
```
